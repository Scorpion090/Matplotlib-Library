# Bar plot word cup 2023.

```python
x = np.random.randint(10,60,(50))
```

```python
n = [43, 30, 25, 31, 50, 32, 43, 30, 40, 48, 39, 26, 54, 38, 11, 35, 44,
       59, 31, 22, 39, 47, 58, 59, 11, 40, 37, 24, 48, 57, 23, 18, 43, 45,
       31, 48, 28, 32, 46, 16, 43, 54, 19, 30, 27, 34, 50, 30, 45, 46]
```

```python
plt.xlabel("Exact value",fontsize = 17,color = "k")
plt.ylabel("Given range",fontsize = 17,color = "k")
plt.title("Histogram",fontsize = 17,color = "k")
plt.hist(n,edgecolor = "r",linewidth = 3,align = "left",cumulative = -1,color = "k")
plt.show()
```

# All the types of histograms - 

__This is a step histogram__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",histtype = "step")
plt.show()
```

__This is a stepfilled histogram__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "Purple",edgecolor = "r",linewidth = 3,align = "left",histtype = "stepfilled")
plt.show()
```

__This is a barstacked histogram__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",histtype = "barstacked")
plt.show()
```

__Convert this histogram verticle to horizontal__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",orientation = "horizontal")
plt.show()
```

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",rwidth = 0.8)
plt.show()
```

__Converting to this table into the log table__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left",log = True)
plt.show()
```

__Converting to this into the like a boxes__

```python
plt.xlabel("Exact value",fontsize = 15,color = "y")
plt.ylabel("Given range",fontsize = 15,color = "y")
plt.title("Histogram plot",fontsize = 16,color = "r")
plt.hist(n,cumulative = -1,color = "k",edgecolor = "r",linewidth = 3,align = "left")
plt.grid()
plt.show()
```

# Stem plot --

```python
x = [2,4,6,8]
y = [1,2,3,4]
plt.xlabel("Range",fontsize = 15,color = "y")
plt.ylabel("Value",fontsize = 15,color = "y")
plt.title("stem plot",fontsize = 15,color = "y")
plt.stem(x,y,linefmt = ":",markerfmt = "y",basefmt = "y",label = "Stem plot",orientation = "vertical")
plt.show()
```
